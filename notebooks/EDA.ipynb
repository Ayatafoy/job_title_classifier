```python
import pandas as pd
from matplotlib import pyplot as plt 
import matplotlib.pylab as pylab
params = {'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large'}

pylab.rcParams.update(params)
pd.set_option('display.max_colwidth', None)
```

```python
jobs_dataset = pd.read_csv('../data/JobsDataset.csv')[['Job Title', 'Query']]
jobs_dataset = jobs_dataset.dropna().drop_duplicates()
```

```python
for title in pd.DataFrame(jobs_dataset['Query'].value_counts()).itertuples():
    print(title.Index)
```

```python
jobs_dataset['Query'].value_counts().plot(kind='bar')
plt.title('Job titles per job category')
plt.show()
```
